```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("kpop_comeback_day_by_day.csv")

# Group and plot average YouTube Views per Artist per Day
avg_views = df.groupby(["Artist", "Day"])["YouTube Views"].mean().reset_index()

sns.lineplot(data=avg_views, x="Day", y="YouTube Views", hue="Artist")
plt.title("Day-by-Day YouTube Views")
plt.show()
```
